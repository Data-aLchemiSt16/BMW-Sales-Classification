```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.rcParams['figure.figsize'] = (12,5)
data = pd.read_csv(r'C:\Users\AVI SHARMA\Documents\CSV-FILES\BMW sales data (2010-2024) (1).csv')
data.head()
```

```python
print("\nShape :", data.shape)

print("\nColumn info:")
print(data.info())

print("\nMissing Values:")
print(data.isnull().sum())
```

VISUALIZATIONS

```python
num_cols = ["Year", "Engine_Size_L", "Mileage_KM", "Price_USD", "Sales_Volume"]
num = len(num_cols)
colors = sns.color_palette("husl", num).as_hex()
for i, col in enumerate(num_cols):
    plt.figure()
    sns.histplot(data[col], kde=True, color=colors[i])
    plt.title(f"Distribution of {col}", fontsize=16)
    plt.show()


cat_cols = ["Model", "Region", "Color", "Fuel_Type", "Transmission", "Sales_Classification"]
for col in cat_cols:
    plt.figure()
    sns.countplot(x=data[col], palette="Set2")
    plt.title(f"Count of {col}", fontsize=16)
    plt.xticks(rotation=45)
    plt.show()


for i, col in enumerate(num_cols):
    plt.figure()
    sns.boxplot(y=data[col], color=colors[i])
    plt.title(f"Boxplot of {col}", fontsize=16)
    plt.show()


plt.figure(figsize=(10,6))
sns.heatmap(data[num_cols].corr(), annot=True, cmap="viridis")
plt.title("Correlation Heatmap - Numerical Features")
plt.show()
```

Advanced visualizations

```python
sns.pairplot(data[["Engine_Size_L","Mileage_KM","Price_USD","Sales_Volume"]],
             diag_kind="kde",
             plot_kws={"alpha":0.6, "color":"purple"})
plt.suptitle("Pairplot Between Numerical Features", y=1.02, fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="Region", y="Sales_Volume", data=data, palette="Set3")
plt.title("Sales Volume Distribution per Region")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.violinplot(data=data, x="Model", y="Price_USD", palette="Spectral")
plt.title("Price Distribution per BMW Model")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
scatter = plt.scatter(data["Engine_Size_L"], data["Price_USD"],
              c=data["Mileage_KM"], cmap="plasma")
plt.colorbar(scatter, label="Mileage KM")
plt.title("Engine Size vs Price (Mileage as Color Gradient)")
plt.xlabel("Engine Size (L)")
plt.ylabel("Price (USD)")
plt.show()
```

```python
plt.figure(figsize=(10,6))
heatmap_data = data.pivot_table(values="Sales_Volume", index="Region", columns="Fuel_Type")
sns.heatmap(heatmap_data, cmap="rainbow", annot=True, fmt=".0f")
plt.title("Heatmap: Fuel Type vs Region (Sales Volume)")
plt.show()
```

```python
data['Sales_Classification'].value_counts().plot(
    kind='pie', autopct='%1.1f%%', colors=colors, explode=[0.05,0.05]
)
plt.title("Sales Classification Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hexbin(data["Mileage_KM"], data["Sales_Volume"], gridsize=20, cmap="viridis")
plt.colorbar(label="Density")
plt.title("Hexbin: Mileage vs Sales Volume")
plt.xlabel("Mileage")
plt.ylabel("Sales Volume")
plt.show()
```

Encoding Data

```python
from sklearn.preprocessing import LabelEncoder
data_encoded = data.copy()

le = LabelEncoder()
data_encoded["Sales_Classification"] = le.fit_transform(data_encoded["Sales_Classification"])
data_encoded = pd.get_dummies(data_encoded, columns=["Model","Region","Color","Fuel_Type","Transmission"], drop_first=True)
data_encoded = data_encoded.astype(int)

print("\n Encoding Completed")
print("New Shape:", data_encoded.shape)
```
